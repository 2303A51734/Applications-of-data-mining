```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/netflix_titles.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 4))
```

```python
plt.hist(df['release_year'], bins=20, edgecolor='black', color='skyblue')
```

```python
plt.title('Release Year Distribution')
```

```python
plt.xlabel('Release Year')
```

```python
plt.ylabel('Frequency')
```

```python
plt.show()
```

```python
plt.figure(figsize=(6, 4))
```

```python
df['type'].value_counts().plot(kind='bar', color='lightcoral')
```

```python
plt.title('Content Type Distribution')
```

```python
plt.xlabel('Type')
```

```python
plt.ylabel('Count')
```

```python
plt.xticks(rotation=0)
```

```python
plt.show()
```

```python
plt.figure(figsize=(6, 4))
```

```python
plt.boxplot(df['release_year'], patch_artist=True, boxprops=dict(facecolor='lightgreen'))
```

```python
plt.title('Box Plot of Release Year')
```

```python
plt.ylabel('Release Year')
```

```python
plt.show()
```

```python
plt.figure(figsize=(8, 4))
```

```python
sns.countplot(x='rating', data=df, palette='Set2', order=df['rating'].value_counts().index)
```

```python
plt.title('Content Rating Count')
```

```python
plt.xlabel('Rating')
```

```python
plt.ylabel('Count')
```

```python
plt.xticks(rotation=45)
```

```python
plt.show()
```

```python
plt.figure(figsize=(6, 4))
```

```python
sns.violinplot(x='type', y='release_year', data=df, palette='muted')
```

```python
plt.title('Release Year Distribution by Content Type')
```

```python
plt.xlabel('Type')
```

```python
plt.ylabel('Release Year')
```

```python
plt.show()
```
